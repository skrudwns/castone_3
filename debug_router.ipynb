```python
from src.graph_flow import build_graph

# ê·¸ëž˜í”„ ë¹Œë“œ
app = build_graph()

# ì‹¤í–‰ ì„¤ì • (ìž¬ê·€ í•œë„ ë„‰ë„‰ížˆ)
config = {"recursion_limit": 50}

print("--- ðŸš€ ì „ì²´ ì›Œí¬í”Œë¡œìš° ì‹œìž‘ ---")

# ìƒíƒœ ìŠ¤íŠ¸ë¦¬ë° ì‹¤í–‰
step_count = 0
try:
    for output in app.stream(initial_state, config=config):
        step_count += 1
        for node_name, state_update in output.items():
            print(f"\n[{step_count}] ðŸ“ í˜„ìž¬ ë…¸ë“œ: {node_name}")
            
            # 1. ì•µì»¤ í™•ì¸
            if "current_anchor" in state_update:
                print(f"   âš“ ì•µì»¤: {state_update['current_anchor']}")
            
            # 2. ì¼ì • ì¶”ê°€ í™•ì¸
            if "itinerary" in state_update:
                itin = state_update['itinerary']
                print(f"   ðŸ“… í˜„ìž¬ ì¼ì • ìˆ˜: {len(itin)}ê°œ")
                if itin:
                    latest = itin[-1]
                    print(f"      ðŸ‘‰ ìµœê·¼ ì¶”ê°€: {latest.get('name')} (Day {latest.get('day')})")

            # 3. ë„êµ¬ í˜¸ì¶œ í™•ì¸ (messagesì— tool_callsê°€ ìžˆëŠ”ì§€)
            if "messages" in state_update:
                last_msg = state_update['messages'][-1]
                if hasattr(last_msg, 'tool_calls') and last_msg.tool_calls:
                    tools = [tc['name'] for tc in last_msg.tool_calls]
                    print(f"   ðŸ› ï¸ ë„êµ¬ í˜¸ì¶œ ì‹œë„: {tools}")
                elif "tool_calls" not in str(last_msg): # ToolMessageê°€ ì•„ë‹Œ ê²½ìš°
                    # í…ìŠ¤íŠ¸ ì‘ë‹µë§Œ í–ˆì„ ë•Œ ë‚´ìš©ì„ ì‚´ì§ ë³´ì—¬ì¤Œ
                    preview = str(last_msg.content)[:50].replace("\n", " ")
                    print(f"   ðŸ’¬ í…ìŠ¤íŠ¸ ì‘ë‹µ: {preview}...")

            # 4. ì¢…ë£Œ ì¡°ê±´ í™•ì¸
            if state_update.get("show_pdf_button"):
                print("\nâœ… PDF ë²„íŠ¼ í™œì„±í™”ë¨! (ì •ìƒ ì¢…ë£Œ ì˜ˆìƒ)")
                
except Exception as e:
    print(f"\nâŒ ì›Œí¬í”Œë¡œìš° ì¤‘ë‹¨ë¨ (ì—ëŸ¬): {e}")
```
